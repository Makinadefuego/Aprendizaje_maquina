```python
from k_means import  Kmeans_classifier 
from liveData import LiveData
from dbscan import Dbscan_classifier

import matplotlib.pyplot as plt
import numpy as np

```

```python
#Se crea una instancia de la clase LiveData
live_data = LiveData( 'Live.csv' )
live_data.clean()
live_data.pemutar()
live_data.visualizarDatos2d()
live_data.visualizarDatos3d()
```

```python
# Se crea una instancia de la clase Kmeans_classifier
kmeans = Kmeans_classifier(live_data.data, live_data.classes, 5) 
kmeans.fit(max_iterations=1000)  # Establece el nmero mximo de iteraciones

# Se obtienen los centroides
centroids = kmeans.centroids

#Se imprime el resultado
print(centroids)

#Se visualiza el resultado
kmeans.visualize_clusters()


#Implementacin del mtodo del codo

def calculate_wcss(data, k):
    wcss_values = []
    for i in range(1, k + 1):
        kmeans = Kmeans_classifier(data, None, i)
        kmeans.fit()
        wcss = 0
        for j in range(len(data)):
            cluster_index = kmeans.predict([data[j]])[0]
            centroid = kmeans.centroids[cluster_index]
            distance = kmeans.euclidean_distance(data[j], centroid)
            wcss += distance ** 2
        wcss_values.append(wcss)
    return wcss_values

# Datos de ejemplo
data = np.random.rand(100, 2)  # Datos aleatorios de 2 dimensiones
max_k = 10  # Valor mximo de k a probar

wcss_values = calculate_wcss(data, max_k)

print(wcss_values)

# Graficar los resultados del mtodo del codo
plt.plot(range(1, max_k + 1), wcss_values, marker='o')
plt.title('Mtodo del Codo')
plt.xlabel('Nmero de Clsteres (k)')
plt.ylabel('WCSS')
plt.show()
```
